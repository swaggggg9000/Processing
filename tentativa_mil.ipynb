```python
from deepface import DeepFace
import os
import shutil


input_folder = '/content/generated_faces'
female_folder = '/content/generated_faces_feminine'
os.makedirs(female_folder, exist_ok=True)


for img_name in os.listdir(input_folder):
    img_path = os.path.join(input_folder, img_name)

    if os.path.isfile(img_path) and img_name.lower().endswith(('.png', '.jpg', '.jpeg')):
        result = DeepFace.analyze(img_path, actions=['gender'], enforce_detection=False)

        if isinstance(result, list):
            result = result[0]

        # Correctly extract the dominant gender
        gender_val = result['dominant_gender'].strip().lower()

        if gender_val == 'woman':
            shutil.copy(img_path, female_folder)
            print(f"Copiando {img_name} -> feminino")

print("Done")
```

```python
import numpy as np
import torch

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Load projected latent
w_path = "/content/projection/projected_w.npz"
w = np.load(w_path)['w']
w = torch.from_numpy(w).to(device)

print("Latent loaded:", w.shape)
```

```python
import copy

G_corrupt = copy.deepcopy(G)  # Create a deep copy of the original generator

with torch.no_grad():
    for name, param in G_corrupt.synthesis.named_parameters():
        # Target convolutional weights in specified layers
        if any(k in name for k in ["b4", "b8", "b16", "b32", "b64","b128"]) and "conv" in name and "weight" in name:
            print(f"Corrupting weights in: {name}")
            param += torch.randn_like(param) * 1.1 # Increased noise multiplier for more distortion
```

**Reasoning**:
Now that the generator weights have been corrupted, I will use both the original and the corrupted generators to create images from the loaded latent vector. Then, I will display them side-by-side to visually compare the original and the proportionally distorted faces.



```python
# Generate images
img_clean = G.synthesis(w, noise_mode='const')
img_corrupt = G_corrupt.synthesis(w, noise_mode='const')

# Helper to show images
import matplotlib.pyplot as plt

def show(img):
    img = (img.permute(0,2,3,1) * 127.5 + 128).clamp(0,255)
    plt.imshow(img[0].cpu().numpy().astype('uint8'))
    plt.axis('off')

# Show side-by-side
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
show(img_clean)
plt.title("Original")
plt.subplot(1,2,2)
show(img_corrupt)
plt.title("Corrupted Early Layer (Increased Noise)")
plt.show()
```

```python
import os
import shutil
from PIL import Image

# Set the number of frames for the video
num_frames = 100

# Create a temporary directory to store the generated image frames
frames_dir = '/content/interpolation_frames'
os.makedirs(frames_dir, exist_ok=True)

print(f"Generating {num_frames} interpolation frames...")

with torch.no_grad():
    for i in range(num_frames):
        # Create a deep copy of the original generator G to G_interp
        G_interp = copy.deepcopy(G)

        # Calculate the current interpolation factor for the frame
        interp_factor = i / (num_frames - 1)

        # Initialize a counter for the corruption_tensors list
        noise_idx = 0

        # Iterate through the named parameters of G_interp.synthesis
        for name, param in G_interp.synthesis.named_parameters():
            # For each convolutional weight in the target layers, apply the corresponding random noise
            if any(f"b{layer_num}" in name for layer_num in [4, 8, 16, 32,64,128]) and "conv" in name and "weight" in name and "affine" not in name:
                # Add the corresponding random noise tensor from corruption_tensors to its value.
                # Scale this noise by the current interpolation factor and the noise_multiplier.
                param.copy_(param + corruption_tensors[noise_idx] * interp_factor * noise_multiplier)
                noise_idx += 1

        # Use the G_interp to synthesize an image from the projected latent vector w.
        img = G_interp.synthesis(w, noise_mode='const')

        # Convert the synthesized image to a suitable format (e.g., NumPy array) and save it
        # The image is scaled from [-1, 1] to [0, 255]
        img_np = (img.permute(0,2,3,1) * 127.5 + 128).clamp(0,255)[0].cpu().numpy().astype('uint8')

        # Save the image to the temporary directory with a sequential filename
        Image.fromarray(img_np, 'RGB').save(os.path.join(frames_dir, f'frame_{i:03d}.png'))

print(f"Successfully generated frames in {frames_dir}/")
```

**Reasoning**:
The interpolation frames have been generated. The next logical step is to combine these frames into a video file to complete the task of creating an MP4 video showing the gradual corruption of a generated face.



```python
import subprocess
import shutil
import os

# Output video file path
output_video_path = '/content/corruption_video.mp4'

# Check if the frames directory exists and contains files
if not os.path.exists(frames_dir) or not any(f.endswith(('.png', '.jpg', '.jpeg')) for f in os.listdir(frames_dir)):
    print(f"Error: Image frames not found in '{frames_dir}'. Please ensure you run the frame generation cell (bd42925a) before compiling the video.")
else:
    # Construct the ffmpeg command to create the video from the frames
    # -framerate: Sets the input frame rate (e.g., 20 frames per second)
    # -i: Specifies the input image sequence. The %03d ensures that frame_000.png, frame_001.png, etc., are correctly processed.
    # -c:v libx264: Sets the video codec to H.264
    # -r: Sets the output frame rate (e.g., 30 frames per second)
    # -pix_fmt yuv420p: Ensures compatibility with most video players
    ffmpeg_command = [
        'ffmpeg',
        '-y', # Overwrite output files without asking
        '-framerate', '200',
        '-i', f'{frames_dir}/frame_%03d.png',
        '-c:v', 'libx264',
        '-r', '30',
        '-pix_fmt', 'yuv420p',
        output_video_path
    ]

    print(f"Creating video: {output_video_path}")

    # Execute the ffmpeg command
    try:
        subprocess.run(ffmpeg_command, check=True, capture_output=True, text=True)
        print("Video creation successful!")
    except subprocess.CalledProcessError as e:
        print(f"Error during video creation: {e}")
        print(f"STDOUT: {e.stdout}")
        print(f"STDERR: {e.stderr}")

# Clean up temporary frames directory if it exists after processing
if os.path.exists(frames_dir):
    shutil.rmtree(frames_dir)
    print(f"Cleaned up temporary directory: {frames_dir}")
else:
    print(f"Temporary directory '{frames_dir}' already removed or never created.")
```
